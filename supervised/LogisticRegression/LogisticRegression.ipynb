# **Logistic Regression**

Logistic regression is a commonly used technique for solving **binary classification problems**. 

---

## **How Logistic Regression Works**

1. **Linear Combination**:  
   - A linear combination (or weighted sum) of the input features is calculated.  

2. **Sigmoid Function**:  
   - The result is passed through the sigmoid function to map it to a value between 0 and 1.  
   - Sigmoid function:  
     \[
     \sigma(z) = \frac{1}{1 + e^{-z}}
     \]
**Visual Summary:**  
![LR3.png](LR1.png)

The output of the sigmoid function is called a logistic, hence the name logistic regression

3. **Probability**:  
   - The output represents the **probability** of the input being classified as "Yes" (or the positive class).  

4. **Loss Function**:  
   - Instead of RMSE, the **Cross-Entropy Loss** function is used to evaluate model performance:  
     \[
     \text{Loss} = - \frac{1}{N} \sum_{i=1}^N \left[ y_i \log(\hat{y}_i) + (1 - y_i) \log(1 - \hat{y}_i) \right]
     \]  
     - \( y_i \): Actual label (0 or 1)  
     - \( \hat{y}_i \): Predicted probability  

---

## **Visual Summary**

- Linear combination of inputs \( z = w_1x_1 + w_2x_2 + \dots + b \)  
- Sigmoid function transforms \( z \) into probabilities: \( \sigma(z) = \frac{1}{1 + e^{-z}} \)  
- Classification is determined by a threshold (e.g., 0.5):  
  - \( \hat{y} = 1 \) if \( \sigma(z) > 0.5 \)  
  - \( \hat{y} = 0 \) otherwise

### Model Training

Parameters (\( \beta \)) are optimized using methods like **Gradient Descent**.

### Multiclass Classification

Logistic regression can be extended to handle multiclass problems (e.g., softmax regression for multiple classes).

---

### Applications of Logistic Regression

- **Medical Diagnosis**: Predicting whether a patient has a disease (Yes/No).  
- **Credit Scoring**: Assessing whether a loan applicant is likely to default.  
- **Marketing**: Determining the likelihood of a customer clicking on an ad.  
- **Manufacturing**: Classifying defective products or predicting equipment failure.  

---

### Advantages

- Simple to implement and interpret.  
- Performs well for linearly separable data.  
- Computationally efficient.  

---

### Limitations

- Assumes linear relationships between features and the log-odds.  
- May underperform with complex relationships or highly imbalanced datasets.  
- Sensitive to outliers.

```python
RawData = pd.read_csv("weatherdata.csv")
RawData.isnull().sum()
```

```python
RawData['Rainfall'].fillna(np.mean(RawData['Rainfall']), inplace = True)
RawData.dropna(inplace = True)

Data = RawData.sort_values(by='Date')
Data.drop(['MinTemp','MaxTemp','Evaporation', 'WindGustDir', 
           'WindGustSpeed', 'WindDir9am', 'WindDir3pm','WindSpeed9am', 
           'WindSpeed3pm', 'Pressure9am','Pressure3pm'], axis=1, inplace=True)
Data.reset_index(drop =True,inplace = True)
Data.index += 1

Data['Humidity'] = Data.apply(lambda row: (row['Humidity9am'] + row['Humidity3pm']) / 2, axis=1)
Data['Cloud'] = Data.apply(lambda row: (row['Cloud9am'] + row['Cloud3pm']) / 2, axis=1)
Data['Temp'] = Data.apply(lambda row: (row['Temp9am'] + row['Temp3pm']) / 2, axis=1)
```
